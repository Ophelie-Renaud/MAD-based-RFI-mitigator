````python
import numpy as np
import matplotlib.pyplot as plt
````

### Initialise an input time-series
#### Quantised Gaussian-distributed random process

````python
nBits = 8
windowLen = 201 # select an odd number for unique median calculation
windowMid = int(windowLen/2)
N = 10000 # number of samples in test time-series
# round N up to the nearest integer multiple of the window length
N = int(np.ceil(N/windowLen)*windowLen)
print(f'Time-series length = {N}')
rndSeq = np.random.randn(N)
# Scale the input to the dynamic range and convert to integers
sigScale = 0.2
scl = sigScale/np.max(np.abs(rndSeq))
xt = np.floor(2**(nBits-1)*scl*rndSeq)
````

````python
plt.plot(xt)
pltH = plt.gcf()
pltH.set_size_inches(10,6)
plt.show()
````

### Add in some lager outliers to test the algorithm against impusive-type RFI

````python
nOutliers = 100
rfiScale = 0.8
outlierIdx = np.floor(N*np.random.rand(nOutliers))
xtRFI = xt.copy()
for i in range(nOutliers):
    xtRFI[int(outlierIdx[i])] = rfiScale*(2**(nBits-1))*np.sign(np.random.randn(1))
````

````python
plt.plot(xtRFI)
pltH = plt.gcf()
pltH.set_size_inches(10,6)
plt.show()
````

### Bock-mode MAD algorithm


````python
nBlocks = int(N/windowLen)
medEstimates = np.zeros(nBlocks)
madEstimates = np.zeros(nBlocks-1)
cumHist = np.zeros(2**nBits)
madCumHist = np.zeros(2**nBits)
sampleRam = np.zeros((2, windowLen))
incDecRom = np.zeros((2**nBits, 2**nBits))

# fill the increment/decrement enable ROM table
for i in range(2**nBits):
    incDecRom[i,i:] = np.ones(2**nBits-i)

# convert input to offset binary (all positive integers)
#xt_ob = 2**(nBits-1) + xt
xt_ob = 2**(nBits-1) + xtRFI
xt_ob[np.where(xt_ob == 2**nBits)] = 2**nBits - 1 # make sure the time-series conforms to nBits offset-binary
bank = 1
bufFull = 0
windowIndex = 0 # RAM banks sample index counter
blockIndex = 0
medianVal = 0

for i in range(N):
    # push a new sampe value into the FIFO
    newSamp = int(xt_ob[i])
    
    # only update median calc's after the first block has been processed
    if (i > 0) & ((i%windowLen) == 0):
        windowIndex = 0
        if bank == 0:
            bank = 1
        else:
            bank = 0
        
        # extract the median value for the current block, reset the cumulative histogram
        medianVal = np.argmax(cumHist>windowMid)
        medEstimates[blockIndex] = medianVal
        cumHist = np.zeros(2**nBits)

        # compute the MAD estimate for the previous sample window
        madVal = np.argmax(madCumHist>windowMid)
        madEstimates[blockIndex] = madVal
        madCumHist = np.zeros(2**nBits)

        # increment the block index
        blockIndex += 1
        
    # push the new sample into the current bank
    sampleRam[bank,windowIndex] = newSamp
    # update the cumulative histogram
    cumHist += incDecRom[newSamp,:]
        
    if blockIndex > 0:
        # subtract the previous window median from the sample in the old bank and compute the absolute deviation
        if bank == 0:
            absDev = np.abs(sampleRam[1, windowIndex] - medianVal)
        else:
            absDev = np.abs(sampleRam[0, windowIndex] - medianVal)
        # update the cumulative histogram for the MAD estimate
        madCumHist += incDecRom[int(absDev),:]

    # increment the ample window counter
    windowIndex += 1
````

### Check calculation using internal Numpy libraries

````python
npMedVals = np.zeros(nBlocks)
npMadVals = np.zeros(nBlocks)
npStdVals = np.zeros(nBlocks)
for i in range(nBlocks):
    sampBuf = xt_ob[i*windowLen:(i+1)*windowLen]
    medVal = np.median(sampBuf)
    npMedVals[i] = medVal - 2**(nBits-1)
    npStdVals[i] = np.std(sampBuf)
    absDev = np.abs(sampBuf - medVal)
    madEst = np.median(absDev)
    npMadVals[i] = madEst
````

````python
plt.plot(medEstimates[:-1]-2**(nBits-1),label='Streaming algorithm')
plt.plot(npMedVals,label='Numpy')
pltH = plt.gcf()
pltH.set_size_inches(10,6)
plt.legend()
plt.title('Block-wise median estimates')
plt.show()
````

````python
plt.plot(madEstimates[1:],label='Streaming algorithm')
plt.plot(npMadVals,label='Numpy')
pltH = plt.gcf()
pltH.set_size_inches(10,6)
plt.legend()
plt.title('Block-wise MAD estimates')
plt.show()
````

### Compare block-wise Standard Deviation ststistic with the equivalent scaled MAD value
For Gaussian-distributed data, the standard deviation is related to the MAD estimate as follows:

$\sigma = k\cdot MAD$

where $k\approx 1.4826$

See https://en.wikipedia.org/wiki/Median_absolute_deviation

````python
plt.plot(madEstimates[1:]*1.4826,label='Scaled MAD, streaming algorithm')
plt.plot(npStdVals,label='Numpy Standard Deviation')
pltH = plt.gcf()
pltH.set_size_inches(10,6)
plt.legend()
plt.title('Block-wise standard deviation vs scaled MAD')
plt.show()
````

````python
tmp = np.ones(windowLen)
madUpper = np.kron(3*madEstimates[1:]*1.4826,tmp)
madLower = np.kron(-3*madEstimates[1:]*1.4826,tmp)
stdUpper = np.kron(3*npStdVals[:-1],tmp)
stdLower = np.kron(-3*npStdVals[:-1],tmp)
plt.plot(xtRFI)
plt.plot(madUpper,label='MAD 3-$\sigma$ upper threshold')
plt.plot(madLower,label='MAD 3-$\sigma$ lower threshold')
plt.plot(stdUpper,label='STD 3-$\sigma$ upper threshold')
plt.plot(stdLower,label='STD 3-$\sigma$ lower threshold')
pltH = plt.gcf()
pltH.set_size_inches(15,8)
plt.legend()
plt.show()
````
